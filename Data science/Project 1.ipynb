```python
import pandas as pd
import seaborn as sns

OriginalData=pd.read_csv('pubg - Dr. Darshan Ingle.csv')
OriginalData.info()
```

```python
OriginalData.describe()
# Here we can see various Statistical results of our data
# On Average person kills is almost 1
# most kills ever recorded are 35
```

```python
OriginalData.columns
```

```python
sns.distplot(OriginalData['matchDuration']);
#By the plot we can observe that most people lasted between 1250 to 1500 and 1750 to 2000 seconds
```

```python
sns.distplot(OriginalData['walkDistance']);
# Most people walked less than 1km and Maximum camped at their own landing positions
```

```python
import  matplotlib.pyplot as plt
# %matplotlib notebook
plt.figure()
plt.subplot(2,1,1)
sns.distplot(OriginalData['matchDuration'].sort_values(),label="Duration");
plt.legend()
plt.subplot(2,1,2)
sns.distplot(OriginalData['walkDistance'].sort_values(),label="Walk");
plt.legend()
```

```python
plt.figure()
plt.subplot(1,2,1)
sns.distplot(OriginalData['matchDuration'].sort_values(),label="matchDuration");
plt.legend()
plt.subplot(1,2,2)
sns.distplot(OriginalData['walkDistance'].sort_values(),label="walkDistance");
plt.legend()
```

```python

Temp1=pd.DataFrame(OriginalData['damageDealt'])
Temp1["kilss"]=OriginalData['kills']
print(Temp1)
Temp2=pd.DataFrame(OriginalData['maxPlace'])
Temp2['numGroups']=OriginalData['numGroups']
sns.pairplot(Temp1)
sns.pairplot(Temp2)
# numGroups vs MaxPlace is Linear
```

```python
OriginalData['matchType'].value_counts()
```

```python
sns.barplot(x=OriginalData['matchType'].sort_values(),y=OriginalData['killPoints'].sort_values())
plt.xlabel("matchType")
plt.ylabel("KillPoints")
plt.xticks(rotation=70)
# it is understood that KillPoints are huge in Normal(Classic) game format
```

```python
sns.barplot(x=OriginalData['matchType'].sort_values(),y=OriginalData['weaponsAcquired'].sort_values())
plt.xlabel("matchType")
plt.ylabel("weaponsAcquired")
plt.xticks(rotation=70)
# weapons acquired incresed drastically in Normal gamemodes
```

```python
sns.boxplot(x=OriginalData['matchType'].sort_values(),y=OriginalData['winPlacePerc'].sort_values())
plt.xlabel("matchType")
plt.ylabel("winPlacePerc")
plt.xticks(rotation=70)
# There are better chances to win with fpp mode
```

```python
sns.boxplot(x=OriginalData['matchType'].sort_values(),y=OriginalData['matchDuration'].sort_values())
plt.xlabel("matchType")
plt.ylabel("matchDuration")
plt.xticks(rotation=70)
#Special modes have short durations, normal squad matches last longer 
```
